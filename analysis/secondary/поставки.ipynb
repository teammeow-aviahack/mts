```python
drop_apr = [
    'Номер строки в заказе/релизе', 
    'Наименование позиции ЕНС',
    'Единица измерения',
    'Полученное количество', 
    'В пути',
    'Дата исполнения заказа/договора', 
    'м3',
    'Дата исполнения заказа/договора*',
    'Дата поставки SWAP Эрикссон',
]

drop_jan = [
    'Номер строки в заказе/релизе',
    'Наименование позиции ЕНС', 
    'Единица измерения',
    'Полученное количество', 
    'В пути', 
    'Длина',
    'Ширина', 
    'Высота', 
    'Дата исполнения заказа/договора', 
    'Группа', 
    'Наименовнание группы',
    'БС',
]

drop_dec = [
    'Дата создания заказа',
    'Номер контракта',
    'Номер строки в заказе/релизе',
    'Наименование позиции ЕНС',
    'Единица измерения',
    'Полученное количество', 
    'В пути', 
    'Длина',
    'Ширина', 
    'Высота',
]
```

```python
data_apr.drop(drop_apr, axis=1, inplace=True)
data_jan.drop(drop_jan, axis=1, inplace=True)
data_dec.drop(drop_dec, axis=1, inplace=True)
```

```python
data_dec['Дата исполнения заказа/договора'] = \
data_dec['Дата исполнения заказа/договора'].apply(lambda x: pd.to_datetime(x, dayfirst=True))
```
