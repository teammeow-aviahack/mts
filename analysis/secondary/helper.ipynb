```python
df_postavki = postavki[['EHC', 'Объём']]
```

```python
df = pd.merge(df_postavki, df_helper_transfer, how="left", on=["EHC"])
```

```python
df = df.dropna().drop_duplicates()
```

```python
df.rename(columns={'Объём': 'real', 'Объем 1 ед. оборудования (куб.м)': 'help'}, inplace=True)
```

```python
df.head()
```

```python
df['delta'] = 0
```

```python
df.info()
```

```python
df['real'] = pd.to_numeric(df['real'], errors='coerce')
df['help'] = pd.to_numeric(df['help'], errors='coerce')
```

```python
import numpy as np
```

```python
df['delta'] = np.abs((df['real'] - df['help'])/df['real'])
```
