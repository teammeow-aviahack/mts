```python
rashod_drop_data = [
    'Тип операции',
    'Описание позиции',
    'Описание организации источника',
    '№ Площадки',
    'ЕИ',
    'ЕИ объема',
]


rashod.drop(rashod_drop_data, axis=1, inplace=True)
```

```python
rashod.groupby('Код организации источника')['Количество операции'].count()
```

```python
rashod['Код организации источника'] = rashod['Код организации источника'].str.replace('К04', '107', regex=True)
rashod['Код организации источника'] = rashod['Код организации источника'].str.replace('К05', '107', regex=True)
rashod['Код организации источника'] = rashod['Код организации источника'].str.replace('101', '107', regex=True)
```

```python
rashod.groupby('Код организации источника')['Количество операции'].count()
```

```python
rashod['year'] = pd.DatetimeIndex(rashod['Дата операции']).year
rashod['month'] = pd.DatetimeIndex(rashod['Дата операции']).month
rashod['day'] = pd.DatetimeIndex(rashod['Дата операции']).day

rashod.drop('Дата операции', axis=1, inplace=True)
```
