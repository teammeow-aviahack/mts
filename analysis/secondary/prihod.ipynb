```python
import pandas as pd
```

```python
prihod = pd.read_excel('../../data/6.Приходные операции_2020-2021.xlsx', skiprows=2)
prihod.head(10)
```

```python
prihod.shape
```

```python
prihod['Код Организации'].unique(), len(prihod['Код Организации'].unique())
```

```python
prihod_drop_data = [
    'Описание позиции',
    'Описание Организации',
    'ЕИ',
    'ЕИ объема',
    'Unnamed: 11',
]


prihod.drop(prihod_drop_data, axis=1, inplace=True)
```

```python
prihod['year'] = pd.DatetimeIndex(prihod['Дата операции']).year
prihod['month'] = pd.DatetimeIndex(prihod['Дата операции']).month
prihod['day'] = pd.DatetimeIndex(prihod['Дата операции']).day

prihod.drop('Дата операции', axis=1, inplace=True)
```

```python
prihod['Код Организации'] = prihod['Код Организации'].str.replace('К04', '107', regex=True)
prihod['Код Организации'] = prihod['Код Организации'].str.replace('К05', '107', regex=True)
prihod['Код Организации'] = prihod['Код Организации'].str.replace('101', '107', regex=True)
```
